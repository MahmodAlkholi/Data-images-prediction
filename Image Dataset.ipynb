# Reading new Data

```python
new_im_df = pd.read_csv(r'E:\AMIT\machine learning\6-SVM\new_images.csv')
new_im_df = new_im_df.iloc[:10000,1:]
new_im_df
```

# splitting data

```python
X = new_im_df.iloc[:, :-1]
y = new_im_df.iloc[:, -1:]
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X,y , test_size=.2 , shuffle=True, random_state=42)
X_train , X_val , y_train , y_val = train_test_split(X_train,y_train , test_size=.2 , shuffle=True, random_state=42)
```

# Normalize

# encoding

```python
from category_encoders import BinaryEncoder
encoder = BinaryEncoder(y)
y_train = encoder.fit_transform(y_train)
y_val = encoder.transform(y_val)
y_test = encoder.transform(y_test)
```

# KNN model

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=50)
```

```python
model.fit(X_train, y_train)
```

**Evaluation**

```python
from sklearn.metrics import accuracy_score

y_tr_pred = model.predict(X_train)
y_val_pred = model.predict(X_val)

tr_acc = accuracy_score(y_train, y_tr_pred)
val_acc = accuracy_score(y_val, y_val_pred)

print(tr_acc)
print(val_acc)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rd_model = RandomForestClassifier()
rd_model.fit(X_train, y_train)
```

# Evaluation

```python
y_tr_predRD = rd_model.predict(X_train)
y_val_predRD = rd_model.predict(X_val)

tr_acc = accuracy_score(y_train, y_tr_predRD)
val_acc = accuracy_score(y_val, y_val_predRD)

print(tr_acc)
print(val_acc)
```
